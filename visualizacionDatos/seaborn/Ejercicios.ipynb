```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#Cargamos el dataset con el que vamos a trabajar
data = sns.load_dataset('penguins')
data.head()
```

## Ejercicio 1
Genera un gráfico de regresión usando las variables 'bill_length_mm' y 'flipper_length_mm'. Implementa este gráfico separado por la variable 'sex'.

```python
sns.lmplot(x = 'bill_length_mm', y= 'flipper_length_mm', data = data, height = 6, hue = 'sex')
```

```python
#También podríamos implemenatarlo usando 'FacetGrid'
graphic = sns.FacetGrid(data = data, col = 'sex', height=6)
graphic.map(sns.regplot, 'bill_length_mm', 'flipper_length_mm')
```

## Ejercicio 2
¿Existe algún valor anómalo (outlier) en alguna variable?


```python
data.columns
```

```python
cols = ['bill_length_mm', 'bill_depth_mm',
       'flipper_length_mm', 'body_mass_g']

i=0
fig, ax = plt.subplots(ncols = 4, figsize = (20, 7))
for c in cols:
    sns.boxplot(data[c], ax = ax[i])
    i+=1
```

Tras observar el grafico podemos ver que no existe ningún valor anomalo o outlier.
Los valores anomalos se mostrarian en este tipo de grafica por encima o por debajo de las barras inferior y superior que estan situadas a los bordes del eje central

## Ejercicio 3
Genera un gráfico que contenga todas las combinaciones de variables. Dicho gráfico debe contener:

- Un histograma en la diagonal
- Un gráfico de puntos encima de la diagonal
- Un gráfico de distribución debajo de la diagonal

```python
graphic = sns.PairGrid(data = data)
graphic.map_diag(sns.histplot) # histograma en la diagonal
graphic.map_upper(sns.scatterplot) #gráfico de puntos encima de la diagonal
graphic.map_lower(sns.kdeplot) #gráfico de distribución debajo de la diagonal
```
