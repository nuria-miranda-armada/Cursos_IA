## Ejercicio 1:
Lee los datos del fichero 'house.csv' en un dataframe.

- ¿Hay alguna habitación con 8 habitaciones?

```python
data = pd.read_csv('house.csv')
data.head()
```

```python
#¿Hay alguna habitación con 8 habitaciones?
bedrooms = data['bedrooms']
8 in bedrooms
```

```python
#¿Cual es el número mínimo y máximo de habitaciones en el DF?
#numero minimo de habitaciones
minimo = bedrooms.min()
#numero maximo de habitacioens
maximo = bedrooms.max()
print(f"numero minimo de habitaciones: {minimo}, numero maximo de habitaciones: {maximo}")
```

```python
#bedrooms
#creamos columna y en ella introcudimos la relacion de precio por planta
data['price_by_floor'] = data['price'] / data['floors']
data.head(2)
```

## Ejecicio 3
Genera un índice multiple con las columnas "street", "city", "statezip" y "country" bajo el nombre "localizacion", y el resto de 
columnas bajo el nombre "caracteristicas". Ordena dichas columnas y genera un nuevo DF que contenga solo las columnas de "localizacion".

```python
data.columns
```

```python
listado = [['caracteristicas','caracteristicas','caracteristicas','caracteristicas','caracteristicas','caracteristicas',
    'caracteristicas','caracteristicas','caracteristicas','caracteristicas','caracteristicas','caracteristicas',
    'caracteristicas','caracteristicas', 'localizacion', 'localizacion', 'localizacion', 'localizacion', 
    'caracteristicas'],
     ['date', 'price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'street', 'city',
       'statezip', 'country', 'price_by_floor']]

tuplas = list(zip(*listado))
tuplas
```

```python
#hago un multi index de las tuplas que acabo de crear
multi_index =  pd.MultiIndex.from_tuples(tuplas)
```

```python
#el data colums pasa a ser el multi_indix asi tendre todas las variables de caracteristicas y todas las variables de localizacion
data.colums = multi_index
data.head()
```

```python
#ordenamos las columnas
#lo ordenamos a nivel caracteristicas por eso asignamos livel 0, axis 1 para indicar la columna
#implace = True para generar un nuevo dataframe
data.sort_index(level = 0, axis = 1, inplace = True)
data.head()
```
